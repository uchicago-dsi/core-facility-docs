```python
import pandas as pd
from config import DATA_DIR

from geocode import geocode_address
```

```python
DATA_PATH = DATA_DIR + "/" + "MPI_Directory_by_Establishment_Number.xlsx"
EXPORT_PATH = DATA_DIR + "/" + "fsis-geocoded.csv"
```

```python
df = pd.read_excel(DATA_PATH)
```

```python
geocoded_rows = []
for idx, row in df[:5].iterrows():
    lat, lng = geocode_address(row.Street, row.City, row.State, str(row.Zip))
    row['lat'] = lat
    row['lng'] = lng
    geocoded_rows.append(row)
```

```python
df_geocoded = pd.DataFrame(geocoded_rows)
df_geocoded.head()
```

```python
df_geocoded.to_csv(EXPORT_PATH)
```
